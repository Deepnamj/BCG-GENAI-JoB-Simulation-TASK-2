```python

file_path = "C:\\Users\\DEEPNA\\Documents\\financial-chatbot-project\\data\\three_companies_financials.xlsx"
df = pd.read_excel(file_path)
df.head()
```

```python
def simple_chatbot(user_query):
    user_query = user_query.lower()  # make it case-insensitive

    if "total revenue" in user_query:
        df_2024 = df[df["Year"] == 2024]
        response = "Hereâ€™s the total revenue for 2024:\n"
        for i, row in df_2024.iterrows():
            response += f"- {row['Company']}: ${row['Total Revenue']} million\n"
        return response

    elif "net income" in user_query:
        df_2023 = df[df["Year"] == 2023]
        df_2024 = df[df["Year"] == 2024]
        response = "Net income change from 2023 to 2024:\n"
        for company in df_2024["Company"].unique():
            val_2023 = df_2023[df_2023["Company"] == company]["Net Income"].values[0]
            val_2024 = df_2024[df_2024["Company"] == company]["Net Income"].values[0]
            change = ((val_2024 - val_2023) / val_2023) * 100
            response += f"- {company}: {change:.2f}% change\n"
        return response

    elif "highest assets" in user_query:
        df_2024 = df[df["Year"] == 2024]
        top = df_2024.loc[df_2024["Total Assets"].idxmax()]
        return f"In 2024, {top['Company']} had the highest total assets (${top['Total Assets']} million)."

    elif "cash flow" in user_query:
        df_2024 = df[df["Year"] == 2024]
        response = "Cash flow from operating activities (2024):\n"
        for i, row in df_2024.iterrows():
            response += f"- {row['Company']}: ${row['Cash Flow from Operating Activities']} million\n"
        return response

    elif "growth" in user_query:
        return "Overall, Microsoft shows consistent growth, Apple remains stable, and Tesla shows fluctuations."

    else:
        return "Sorry, I can only answer specific financial queries like total revenue, net income, assets, or cash flow."
```

```python
while True:
    query = input("Ask the chatbot (or type 'exit' to quit): ")
    if query.lower() == "exit":
        print("Goodbye ðŸ‘‹")
        break
    print(simple_chatbot(query))
```
